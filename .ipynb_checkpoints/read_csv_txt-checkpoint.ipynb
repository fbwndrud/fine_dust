# 데이터 읽고 데이터 프레임에 저장

```python
from os import listdir
from os.path import isfile, join
import pandas as pd

def read_csv_infolder(source_folder):
    csv_files = [f for f in listdir(source_folder) if isfile(join(source_folder, f))]
    new_list = []
    for i,v in enumerate(csv_files):
        new_dict ={}
        if 'txt' in v:
            new_dict['sub'] = pd.read_csv(r'{}\{}'.format(source_folder,v),sep = '\t',encoding = 'utf-8')
            new_dict['title'] = v
            new_list.append(new_dict)
        elif 'csv' in v:
            new_dict['sub'] = pd.read_csv(r'{}\{}'.format(source_folder,v),sep = '|',encoding = 'utf-8')
            new_dict['title'] = v
            new_list.append(new_dict)
    new_df = pd.DataFrame(new_list)
    return new_df
```

```python
def year(df):
    index = df.MCT_CAT_CD.value_counts().index
    upjong = [df.append(card_spending_df['sub'][0].loc[card_spending_df['sub'][0]['MCT_CAT_CD']==j,:]) for j in index]
    for i in range(len(upjong)):
        label = []
        label.append(df[i]['MCT_CAT_CD'].iloc[0])
        upjong[i] = upjong[i].groupby(df[i].STD_DD).sum()
        upjong[i] = upjong[i].sort_values('STD_DD')
        upjong[i] = upjong[i][['USE_AMT']]
        upjong[i] = upjong[i].rename(columns = {"USE_AMT":label[0]})
    for i in range(1,len(upjong)):
        upjong[0] = upjong[0].merge(df[i],how='outer',on='STD_DD')
    upjong = upjong[0]
    
    return upjong
```
